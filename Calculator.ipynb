```python
def add(num1,num2):
    add = num1 + num2
    return add

def sub(num1,num2):
    sub = num1 - num2
    return sub
    
def mul(num1,num2):
    mul = num1 * num2
    mul = num2 * num1
    return mul
    
def div(num1,num2):
    div = num1 / num2
    return div

def int_div(num1,num2):
    int_div = num1 // num2
    return int_div

def expo(num1,num2):
    expo = num1 ** num2
    return expo

def mod(num1,num2):
    mod = num1 % num2
    return mod

def logarithm(num1,num2):
    logarithm = math.log(num1,num2)
    return logarithm

def userinput(num1, num2):
    return "You not Choose any operation\n"

switches = {
        '+' : add,
        '-' : sub,
        '%' : mod,
        '*' : mul,
        '/' : div,
        '//': int_div,
        '**': expo,
        'l': logarithm
}
#   get is returning the same value from switch,that we enter when choosing operator
def switch(operator,num1,num2):
    return switches.get(operator,userinput)(num1,num2)

if __name__=='__main__':   
    while True:
        print("\nWelcome To Mathe World")
        print("Choose the operation to be performed \n")
        print("Addition + ")
        print("Subtraction - ")
        print("Multiplication * ")
        print("Division / ")
        print("Integer Division // ")
        print("Modulus % ")
        print("Exponential ** ")
        try:
            operator = input("Enter the operator : ")
            num1 = int(input("Enter First Value: "))
            num2 = int(input("Enter Second Value: "))
        except:
            print("Invalid Credinals \n")
        try:
            print("Result = ",switch(operator,num1,num2))
        except:
            print("\nError found...!!!!\n")
        try:
            repeat = int(input("Want to continue? (1/0) : "))
        except:
            print("Binary is acceptable.....")
        try:
            if repeat == 0:
                print("\nThanks for using....:)")
                break
            elif repeat <= 1:
                continue
            else:
                print("Wrong Credinals")
        except:
                print("Invalid Credinals......")
```
